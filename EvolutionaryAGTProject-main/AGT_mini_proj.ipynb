```python
dove = {'dove':[0.5,0.5],'hawk':[0.5,0.5]}          #initialising behavioural response of each species
hawk = {'dove':[1,-0.5],'hawk':[1,-0.5]}

behaviour = {'dove':dove,'hawk':hawk}

truedove = {'looks':'dove','is':'dove'}             #initiaising genes with which the looks and behaviour will be defined
imposterdove = {'looks':'dove','is':'hawk'}
truehawk = {'looks':'hawk','is':'hawk'}
imposterhawk = {'looks':'hawk','is':'dove'}

players = {'truedove':truedove,'imposterdove':imposterdove,'truehawk':truehawk,'imposterhawk':imposterhawk}   #initialising players
```

```python
def sum_food(a,b):      #helper function which returns utility on the basis of first player's nature and
  return (a[0]+b[1])        #second players looks

def get_food(a,b):      #function to give utility of the players
  
  return sum_food(behaviour[players[a]['is']][players[b]['looks']],behaviour[players[b]['is']][players[a]['looks']])

def simulation(days,populations):           #function to simulate over a given amount of food
  
  pop_arr = {'truedove':[],'imposterdove':[],'truehawk':[],'imposterhawk':[]}         #initialising populationd trend arrays
  D = 5e4 # number of food

  for i in range(days):         #running for each day

    for p1 in players:          #loop for calculating new population for each player each day

      new_pop = 0
      for p2 in players:
        new_pop += populations[p1]*(populations[p2]/D)*get_food(p1,p2)      #each player's utility*probability with each player

      new_pop += populations[p1]*(1 - sum(populations.values())/D)*2        #each player if it finds no competition

      new_pop = round(new_pop)
      pop_arr[p1].append(new_pop)
    
    for p in players:
      populations[p] = pop_arr[p][-1]             #updating current population with new population

  for i in players:                           #plottting for each player
    plt.plot(pop_arr[i],label = i)

  plt.legend()
  plt.show()
```

```python
for i in players:             #utilities for each player against each player
  for j in players:
    print(i,'vs.',j,' = ',get_food(i,j))
```

```python
populations = {'truedove':3e4,'imposterdove':2e3,'truehawk':3e3,'imposterhawk':1e3}
simulation(40,populations)
```
